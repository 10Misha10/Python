```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(0, n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

from tkinter import *

def get_arr(str):
    arr = []
    for char in str.split(sep=','):
        if char.isdigit():
            arr.append(int(char))
    return arr

def clicked():
    arr = get_arr(txt.get())
    bubble_sort(arr)
    numbers_as_strings = [str(num) for num in arr]
    result = ','.join(numbers_as_strings)
    lbl.config(text=result)
    
def clicked2():
    check = False
    for char in lbl.cget("text").split(sep=','):
        if int(char) == int(txt2.get()):
            lbl2.config(text=f"Число {txt2.get()} найдено в списке")
            check = True
            break
    if not check:
        lbl2.config(text=f"Число {txt2.get()} не найдено в списке")

window = Tk()
window.title("Task1")
window.geometry('300x150')

txt = Entry(window, width=15)
txt.place(x=5, y=5)

btn = Button(window, text="Сортировать", command=clicked)
btn.place(x=120, y=0)

lbl = Label(window)
lbl.place(x=120, y=30)

txt2 = Entry(window, width=15)
txt2.place(x=5, y=70)

btn2 = Button(window, text="Поиск", command=clicked2)
btn2.place(x=120, y=65)

lbl2 = Label(window)
lbl2.place(x=120, y=95)

window.mainloop()
            
```

```python
class OpenAddressingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return len(key) % self.size

    def insert(self, key, value):
        hash_key = self.hash_function(key)
        
        while self.table[hash_key] is not None:
            hash_key = (hash_key + 1) % self.size
        
        self.table[hash_key] = (key, value)
        
    def get(self, key):
        hash_key = self.hash_function(key)
        
        while self.table[hash_key] is not None:
            stored_key, value = self.table[hash_key]
            if stored_key == key:
                return value
            hash_key = (hash_key + 1) % self.size
        
        return None
    
    def remove(self, key):
        hash_key = self.hash_function(key)
        
        while self.table[hash_key] is not None:
            stored_key, _ = self.table[hash_key]
            if stored_key == key:
                
                self.table[hash_key] = None
                return
            hash_key = (hash_key + 1) % self.size

hash_table = OpenAddressingHashTable(10)

hash_table.insert("apple", 5)
hash_table.insert("banana", 10)
hash_table.insert("cherry", 15)

print(hash_table.get("apple"))
print(hash_table.get("banana"))  
print(hash_table.get("cherry")) 

hash_table.remove("banana")
hash_table.remove("apple")

print(hash_table.get("apple"))
print(hash_table.get("banana")) 

```
